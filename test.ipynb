```python

url = 'https://raw.githubusercontent.com/xichennn/paper1/main/training_data_1000.csv'
```

```python
data = pd.read_csv(url,header=None)
data.head()
# plt.plot(range(len(data.iloc[1,20:200])),data.iloc[1,20:200])
train, test = train_test_split(data, test_size = 0.4, random_state = 2021)
```

```python
# data_all = data.to_numpy()
# X = data_all[:,1:19]
# Y = data_all[:,20:200]
X_train = train.iloc[:,1:19]
# Y_train = train.iloc[:,20:200]
Y_train = train.iloc[:,20:21]
X_test = test.iloc[:,1:19]
Y_test = test.iloc[:,20:200]
```

Fit GP model


> Dimension reduction on Y



```python
kernel = GPy.kern.ExpQuad(input_dim=X_train.shape[1],ARD=True,variance=1)
```

```python
model = GPy.models.GPRegression(X_train,Y_train,kernel)
```

```python

display(model)
```

```python
model.optimize(messages=True)
```

```python
model.optimize_restarts(num_restarts = 10)
```
